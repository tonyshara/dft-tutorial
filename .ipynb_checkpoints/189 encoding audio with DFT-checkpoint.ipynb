```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import scipy.cluster
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
%matplotlib inline

import IPython.display as ipd
from scipy import signal, linalg
from scipy.io import wavfile

import glob
import pyaudio
import wave
```

Once we have decided on a word that we would like to use as our password, we will collect data.  This data will consist of multiple sample of us saying the password.  You can chose the number of samples (NUM_SAMPLES) you would like to take and the sampling rate (fs)

```python


NUM_SAMPLES = 5
chunk = 1024  # Record in chunks of 1024 samples
sample_format = pyaudio.paInt16  # 16 bits per sample
channels = 1
fs = 22500  # Record at 44100 samples per second
seconds = 2
filename = "tony2"

p = pyaudio.PyAudio()  # Create an interface to PortAudio

print('Recording')

stream = p.open(format=sample_format,
                channels=channels,
                rate=fs,
                frames_per_buffer=chunk,
                input=True)
samples = []
frames = []  # Initialize array to store frames

# Store data in chunks for 3 seconds
for dp in range(NUM_SAMPLES):
    print("sampling data point \n", dp)
    for i in range(0, int(fs / chunk * seconds)):
        data = stream.read(chunk)
        frames.append(data)
    samples.append(frames)
    frames = []
# Stop and close the stream 
stream.stop_stream()
stream.close()
# Terminate the PortAudio interface
p.terminate()

print('Finished recording')

# Save the recorded data as a WAV file
print("saving recordings")
for dp in range(NUM_SAMPLES):
    wf = wave.open("wavs/"+filename+"_"+str(dp)+".wav", 'wb')
    wf.setnchannels(channels)
    wf.setsampwidth(p.get_sample_size(sample_format))
    wf.setframerate(fs)
    wf.writeframes(b''.join(samples[dp]))
    wf.close()
print("finished saving recordings")
```

Now we have collected our data, where each datapoint is in the form of a .wav file, we can construct our data matrix.  To do this, we first need to get each .wav file into a usable format.
```python
data_points = []
print(glob.glob("wavs/*.wav"))
for fn in glob.glob("wavs/*.wav"):
    fs, data = wavfile.read(fn)
    print(data.shape)
    data_points.append(data)
data_matrix = np.array(data_points)
print(data_matrix.shape)

plt.plot(data_points[0].T)
plt.show()
plt.plot(data_points[1].T)
plt.show()
plt.plot(data_points[2].T)
plt.show()
plt.plot(data_points[3].T)
plt.show()
plt.plot(data_points[4].T)
plt.show()
plt.plot(data_matrix.T)
plt.show()
```

```python
def get_snippets(data, length, pre_length, thres):
    data_out = np.zeros((np.shape(data)[0], length))
    
    for rnum, row in enumerate(data):
        # Find the threshold
        row_thres = thres*np.max(row)

        # Figure out when interesting snippet starts
        block = pre_length
        while (row[block] < row_thres):
            block = block + 1
        block = min(block, 172 - length)
        data_out[rnum,:] = row[block-pre_length:block-pre_length+length]
        
        # Normalization
        data_out[rnum,:] = data_out[rnum,:] / np.sum(data_out[rnum,:])
        
    return data_out
```
